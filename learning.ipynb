## Get models from ollama

```python
import ollama_models_enum
```

```python
result= ollama_models_enum.fetch_ollama_models()
print(result)
```

```python
result= ollama_models_enum.create_ollama_models_enum()
print(result)
print("-"*20)
for item in result:
    print(item)
```

```python
result= ollama_models_enum.get_ollama_models_enum()
print(result)
print("-"*20)
for item in result:
    print(item.value)
```

```python
ollama_models_enum.clear_models_cache() # اگر مدل جدید دانلود شد باید کش را پاک کرد
result = ollama_models_enum.get_ollama_models_enum()

print("نمایش کامل (پیش‌فرض):")
for item in result:
    print(item)  # OllamaModels.TRANSLATEGEMMA_FAIR_LATEST

print("\n" + "="*50)
print("فقط نام:")
for item in result:
    print(item.name)  # TRANSLATEGEMMA_FAIR_LATEST

print("\n" + "="*50)
print("فقط مقدار (نام مدل):")
for item in result:
    print(item.value)  # translategemma-fair:latest

print("\n" + "="*50)
print("فرمت سفارشی:")
for item in result:
    print(f"{item.name:30} -> {item.value}")
```
